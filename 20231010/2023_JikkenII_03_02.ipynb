## アヤメのデータセット
データを回収したり作ったり確認したりする．

````python
# -*- coding: utf-8 -*-
from sklearn import datasets
import numpy as np

# アヤメのデータをロードし、変数irisに格納
iris = datasets.load_iris()

# 特徴量のセットを変数Xに、ターゲットを変数yに格納
X = iris.data
y = iris.target

# Xとyをそのまま表示
print(X)
print(y)

# Xとyの次元を表示。それぞれ(150, 4)と(150,)となる。
# サンプル数150, 特徴量の次元4, という意味
print(X.shape)
print(y.shape)

# サンプル数と特徴量の次元の取り出し方法
(n_samples, n_features) = X.shape
print('サンプル数: {0}'.format(n_samples))
print('特徴量の次元: {0}'.format(n_features))

# クラス数の取り出し方法
n_classes = len(np.unique(y))
print('クラス数: {0}'.format(n_classes))
````

それに加えてそれらのデータセットのデータを扱うための方法も一緒に持っている．  

この例のソースコードで関係しているところを抜き出すと

```
from sklearn import datasets

# アヤメのデータをロードし、変数irisに格納
iris = datasets.load_iris()
```

この部分で `sklearn` パッケージから `datasets` モジュールを `import` して，そのクラスを変数 `iris` に格納する．  
これで，`iris` は `sklearn` の `datasets` モジュールで定義されたクラスを指している形となる．  
これにより，`iris` はモジュール内で定義されたクラスに従って，インスタンス変数 (`iris.data`，`iris.target`) やインスタンス関数 (`load_iris()`) を利用できるようになる．  

ということで，アヤメのデータセット本体を読み込んでいる．

それからデータとターゲットをそれぞれ変数 `X` と変数 `y` に読み込んでいる．  
データを `X`，ターゲットを `y` とするのは慣例的によくやる．

```
# Xとyをそのまま表示
print(X)
```

`X` は特徴量が 4 個なので 150 $\times$ 4 となり，ターゲット `y` は 150 $\times$ 1 となる．

```
# サンプル数と特徴量の次元の取り出し方法
(n_samples, n_features) = X.shape
print('サンプル数: {0}'.format(n_samples))

```
`shape` で返ってくるタプルの値 `(150, 4)` のそれぞれを変数に入れて表示している．

サンプル数や特徴量の次元の表示で，フォーマット付き文字列の表示方法を使っています．( `C` 言語で `printf()` 関数ってありましたよね？)

Python でもフォーマット付きの表示方法があります．  
大変めんどくさいことに何種類かあって，Python のバージョンで使えたり使えなかったり推奨してたりしてなかったりしてます．  
今回使っている方法は現在のバージョンで使える方法です．  

`'表示したい文字列 {}'.format(表示したい変数やリテラル)`  

という形で書きます．画面で表示されるのはクォーテーションの中身で，その中の波カッコ (`{}`) が `format()` のカッコ内の変数やリテラル (定数) に置き換えられます．
```
'hogehoge {0}, {1}, funifuni'.format(変数0, 変数1)'
```
のように，表示部分に `{}` がいくつかあったら，特に指定がなければ先頭から順番に入れていかれて，「hogehoge <変数 0 の中身>，<変数 1 の中身>，funifuni」のように表示されます．  

ここでは
```
print('サンプル数: {0}'.format(n_samples))
print('特徴量の次元: {0}'.format(n_features))
```
という形でサンプル数と特徴量の次元を表示しています．

```
# クラス数の取り出し方法
n_classes = len(np.unique(y))
print('クラス数: {0}'.format(n_classes))
```
データは `numpy.ndarray` であり，NumPy の形式の配列．  
NumPy の `np.unique()` モジュールを使って，ある NumPy 配列 (`ndarray`) の中の
データの重複なしの個数を取ってくることができる．  
ここでいう "unique (ユニーク)" は "ただ一つの" ということ．  
クラス数は 3 個，各クラスにそれぞれ 0, 1, 2 が割り当てられているのは
前回もやった通り．  
ということで要素 3 個の NumPy 配列が返ってくるので，その要素数を数えて
3 クラスということがわかる．
