## 練習 2

小手調べ．

円周率 $\pi$ を乱数を使って求めよう．

- 求め方
  - 2 $\times$ 2 の正方形と直径 2 の円を考える．
  - $xy$ 座標を一様乱数で生成して，発生させた点の数と円の内側にできた点の数の比を取る．
  - 正方形の面積 $2 \times 2 = 4$ と円の面積 $\pi \times {1}^{2} = \pi$ とその点の数の比率を合わせて考えて，$\pi$ を求める．

- 乱数で生成する点の個数とランダム性による各試行での値の変化を考えて，個数などを変化させて繰り返し試行して，その平均値を求めることが望ましい．
- numpy.random.rand が $0$ 以上 $1$ 未満の範囲での乱数を生成するので，それを考えて第 1 象限の $1/4$ の円と正方形から考えてもいい．

```python

all = 0
allIn = 0

for counter in range(10):
    print("time", counter)
    data = np.random.rand(10000, 2)
#    dis = data[data]
    dis = [(True if x[0]*x[0]+x[1]*x[1] < 1 else False) for x in data]
    inCircle = len([0 for x in dis if x])
    gen = len(data)
    print("generate = ", gen, ", in-circle = ", inCircle, sep="")
    print(inCircle / gen * 4)
    print()
    all += gen
    allIn += inCircle
print("report")
print("generate = ", all, ", in-circle = ", allIn, sep="")
print(allIn / all * 4)

```
